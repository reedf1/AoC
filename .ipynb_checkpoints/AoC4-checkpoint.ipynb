```python
# Advent of Code Day 4!
# By Reed Roberts
# December 4nd 2020

input_file_path = "input/input4-2.txt";

input_file = open(input_file_path, "r");

passport_string = '';

for line in input_file:
    passport_string += line;
    
passport_list = passport_string.split('\n\n')

cleaned_passport_list = [];

for line in passport_list:
    cleaned_passport_list.append(line.replace('\n',' ').split(' '));

del (cleaned_passport_list[-1])[-1];

cleaned_passport_list
```

```python
# The automatic passport scanners are slow because they're having trouble detecting which passports have all required fields. The expected fields are as follows:
# 
# byr (Birth Year)
# iyr (Issue Year)
# eyr (Expiration Year)
# hgt (Height)
# hcl (Hair Color)
# ecl (Eye Color)
# pid (Passport ID)
# cid (Country ID)

import re;

valid_fields = ['byr','iyr','eyr','hgt','hcl','ecl','pid','cid'];
hacked_valid_fields = ['byr','iyr','eyr','hgt','hcl','ecl','pid'];

valid_fields.sort()
hacked_valid_fields.sort()

valid_passports = [];
for passport in cleaned_passport_list:
    field_list = [];
    
    for field in passport:
        field_list.append(field.split(':')[0]);
    
    field_list.sort();
    if (valid_fields == field_list):
        valid_passports.append(passport);
    elif(hacked_valid_fields == field_list):
        valid_passports.append(passport);
        

valid_ecl = ['amb', 'blu', 'brn','gry', 'grn', 'hzl', 'oth'];


def generate_passport_dict( passport ):
    passport_dict = {
    'byr' : '',
    'iyr' : '',
    'eyr' : '',
    'hgt' : '',
    'hcl' : '',
    'ecl' : '',
    'pid' : '',
    'cid' : ''
    }
    for field in passport:
        split_field = field.split(':');
        field_key = split_field[0];
        field_value = split_field[1];
        passport_dict[field_key] = field_value;
    return passport_dict

print(len(valid_passports));

checked_valid_passports = [];
for passport in valid_passports:
    passport_dict = generate_passport_dict(passport);
    
    valid_passport = True;
    
    # check birth year
    byr = passport_dict['byr'];
    if(not ( len(byr) == 4 and (1920 <= int(byr) <= 2002))):
        valid_passport = False;
        
    # check issue year
    iyr = passport_dict['iyr'];
    if(not ( len(iyr) == 4 and (2010 <= int(iyr) <= 2020))):
        valid_passport = False;
    
    # check expiration year
    eyr = passport_dict['eyr'];
    if(not ( len(eyr) == 4 and (2020 <= int(eyr) <= 2030))):
        valid_passport = False;
        
    # check height
    height = passport_dict['hgt'];
    if ('in' in height):
        height_int = int(height[:-2]);
        
        if(not (59 <= height_int <= 76)):
            valid_passport = False;
            
    elif ('cm' in height):
        height_int = int(height[:-2]);
        
        if(not (150 <= height_int <= 193)):
            valid_passport = False;
    else:
        valid_passport = False;
    
    # check hair color
    hair_color = passport_dict['hcl'];
    
    if (not re.match(r'#[a-f0-9]{6}',hair_color) ):
        valid_passport = False;
        
    # check eye color
    eye_color = passport_dict['ecl'];
    
    if (not (eye_color in valid_ecl)):
        valid_passport = False;

    # check pid
    pid = passport_dict['pid'];
    
    if (not (len(pid) == 9 and pid.isdigit())):
        valid_passport = False;
        
    if(valid_passport):
        checked_valid_passports.append(passport);
        
print(len(checked_valid_passports))
```
