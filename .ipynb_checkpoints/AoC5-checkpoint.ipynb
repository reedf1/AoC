```python
# Advent of Code Day 5!
# By Reed Roberts
# December 5th 2020

input_file_path = "input/input5.txt";

input_file = open(input_file_path, "r");

formatted_list = [];

for line in input_file:
    formatted_list.append(line.rstrip());
    
print(formatted_list);
```

```python
import numpy as np

def row_partition(boarding_code, seating_rows):
    if(len(seating_rows) == 1):
        return boarding_code,seating_rows
    
    boarding_char = boarding_code[0];
    mid_point = int(len(seating_rows)/2);
    if (boarding_char == 'F'):
        return row_partition(boarding_code[1:], seating_rows[:mid_point])
    elif (boarding_char == 'B'):
        return row_partition(boarding_code[1:], seating_rows[mid_point:])

def column_partition(column_code, seating_columns):
    if(len(seating_columns) == 1):
        return seating_columns
    
    boarding_char = column_code[0];
    mid_point = int(len(seating_columns)/2);
    if (boarding_char == 'L'):
        return column_partition(column_code[1:], seating_columns[:mid_point])
    elif (boarding_char == 'R'):
        return column_partition(column_code[1:], seating_columns[mid_point:])

seating_rows = list(range(128))
seating_columns = list(range(8))
list_of_seats = [];
for boarding_code in formatted_list:
    column_code, row = row_partition(boarding_code,seating_rows)
    column = column_partition(column_code, seating_columns);
    list_of_seats.append((row[0],column[0]));


seating_array = [];
for row in seating_rows:
    seating_array.append(len(seating_columns)* [0]);
    
for seat in list_of_seats:
    seating_array[seat[0]][seat[1]] = 'x';
    
    
    
np.set_printoptions(threshold=np.inf)
print(np.array(seating_array));
```
