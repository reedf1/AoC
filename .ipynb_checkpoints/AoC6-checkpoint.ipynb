```python
# Advent of Code Day 6!
# By Reed Roberts
# December 6th 2020

input_file_path = "input/input6.txt";

input_file = open(input_file_path, "r");

passport_string = '';

for line in input_file:
    passport_string += line;
    
passport_list = passport_string.split('\n\n')

cleaned_passport_list = [];

for line in passport_list:
    cleaned_passport_list.append(line.replace('\n',' ').split(' '));

del (cleaned_passport_list[-1])[-1];

customs_declaration_list = cleaned_passport_list;
customs_declaration_list
```

```python
total_yes_questions = 0;
for group in customs_declaration_list:
    answers_sets = []
    for answers in group:
        answers_set = set();
        for answer in answers:
            answers_set.add(answer);
        answers_sets.append(answers_set);    
    print(answers_sets);
            
print(total_yes_questions)
```
