```python
from tqdm import tqdm
import copy

def find_seat(idx,idy,input_list):
    if(idx < 0 or idy < 0 or idy >= len(input_list) or idx >= len(input_list[0])):
        return None
    
    return input_list[idy][idx];

def count_occupied_surrounding(idx,idy,input_list):
    seat_count = 0;
    
    n_look = list(range(idy-1, -1,-1));
    s_look = list(range(idy+1, len(input_list)));
    e_look = list(range(idx+1, len(input_list[0])));
    w_look = list(range(idx-1, -1,-1));
    
    
    # look north
    for y in n_look:
        seat = find_seat(idx,y,input_list);
        if (seat == '#'):
            seat_count += 1;
            break;
        elif(seat == 'L'):
            break;
            
    # look south;
    for y in s_look:
        seat = find_seat(idx,y,input_list);
        if (seat == '#'):
            seat_count += 1;
            break;
        elif(seat == 'L'):
            break;
            
    # look west
    for x in w_look:
        seat = find_seat(x,idy,input_list);
        if (seat == '#'):
            seat_count += 1;
            break;
        elif(seat == 'L'):
            break;
    
    # look east
    for x in e_look:
        seat = find_seat(x,idy,input_list);
        if (seat == '#'):
            seat_count += 1;
            break;
        elif(seat == 'L'):
            break;
            
    # look ne
    for x,y in zip(e_look,n_look):
        seat = find_seat(x,y,input_list);
        if (seat == '#'):
            seat_count += 1;
            break;
        elif(seat == 'L'):
            break;
    
    # look nw
    for x,y in zip(w_look,n_look):
        seat = find_seat(x,y,input_list);
        if (seat == '#'):
            seat_count += 1;
            break;
        elif(seat == 'L'):
            break;
            
    # look se
    for x,y in zip(e_look,s_look):
        seat = find_seat(x,y,input_list);
        if (seat == '#'):
            seat_count += 1;
            break;
        elif(seat == 'L'):
            break;
            
    # look sw
    for x,y in zip(w_look,s_look):
        seat = find_seat(x,y,input_list);
        if (seat == '#'):
            seat_count += 1;
            break;
        elif(seat == 'L'):
            break;
    
    
    return seat_count;
    
def perform_simulation(simulation_input):
    simulated_output = copy.deepcopy(simulation_input);
    
    # loop through every spot
    for idy,row in enumerate(simulation_input):
        for idx,column in enumerate(simulation_input[idy]):
            if (find_seat(idx,idy,simulation_input) == 'L'): 
                # seat is empty
                if(count_occupied_surrounding(idx,idy,simulation_input) == 0):
                    #seat is empty so occupy it
                    simulated_output[idy][idx] = '#';
            elif(find_seat(idx,idy,simulation_input) == '#'):
                #seat is occupied
                if(count_occupied_surrounding(idx,idy,simulation_input) >= 5):
                    simulated_output[idy][idx] = 'L';
                
    return simulated_output;

def count_occupied_seats(input_list):
    occupied_seats = 0;
    for idy,row in enumerate(input_list):
        for idx,column in enumerate(input_list[idy]):
            if (find_seat(idx,idy,input_list) == '#'):
                occupied_seats += 1;
                
    return occupied_seats;

# run simulation!
loop_condition = True;
current_simulation = formatted_list;
while(loop_condition):
    current_simulation = perform_simulation(current_simulation);
    print(count_occupied_seats(current_simulation))
```
