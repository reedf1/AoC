```python
import math

# Starting at the top-left corner of your map and following a slope of right 3 and down 1, how many trees would you encounter?

def find_tree_count(width_movement,height_movement,tree_map):
    slope_height = len(tree_map);
    slope_pattern_width = len(tree_map[0]); #this is the width of the slope pattern

    steps_to_bottom = math.ceil(slope_height / height_movement);

    # quickndirty
    slope_width = steps_to_bottom*width_movement;  #this is the required width to reach the bottom in this case

    number_of_pattern_repeats = math.ceil(slope_width/slope_pattern_width);

    full_slope = [];
    for line in tree_map:
        full_slope.append(line*number_of_pattern_repeats)

    x_position = 0;
    y_position = 0;
    tree_count = 0;

    for step in range(steps_to_bottom):
        if(full_slope[y_position][x_position] == '#'):
            tree_count += 1;
        x_position += width_movement;
        y_position += height_movement;

    return tree_count

tree1 = find_tree_count(1,1,formatted_list);
tree2 = find_tree_count(3,1,formatted_list);
tree3 = find_tree_count(5,1,formatted_list);
tree4 = find_tree_count(7,1,formatted_list);
tree5 = find_tree_count(1,2,formatted_list);
print(tree1, " ", tree2, " ",tree3, " ", tree4, " ",tree5)

tree1*tree2*tree3*tree4*tree5

# Right 1, down 1.
# Right 3, down 1. (This is the slope you already checked.)
# Right 5, down 1.
# Right 7, down 1.
# Right 1, down 2.
# In the above example, these slopes would find 2, 7, 3, 4, and 2 tree(s) respectively; multiplied together, these produce the answer 336.
# 
# What do you get if you multiply together the number of trees encountered on each of the listed slopes?

```
